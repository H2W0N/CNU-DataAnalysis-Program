```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw = pd.read_csv("/content/drive/MyDrive/school/CNU_2022_data/data/학생건강검사 결과분석 rawdata_서울_2015_20200114.csv", 
                         encoding = 'cp949')
data_raw.head()
```

```python
# 데이터 탐색

data_raw.shape
```

```python
data_raw.describe()
```

```python
# 키와 몸무게 데이터만 저장

data = data_raw[['키', '몸무게']]
data.head()
```

```python
# 결측치 확인!

data.isnull().sum()
```

```python
# 결측치 제거

data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
# 키와 몸무게를 각각 따로 저장한 후 scatter 시키기.

height = data['키']
weight = data['몸무게']

plt.scatter(weight, height, s = 3)
plt.title("Korean teen health data", fontsize = 20)
plt.xlabel("weight")
plt.ylabel("height")
plt.show()
```

```python
# 임의의 n개의 데이터를 이용하고 싶다! -> random
# np.random.randint(시작, 끝, 몇개뽑을거니?)

np.random.randint(1, 10, 5)
```

```python
# 난수를 고정하기.

np.random.seed(16)
np.random.randint(1,10,5)
```

```python
data.shape[0]
```

```python
# 임의의 데이터 100개를 뽑을 index 생성
select = np.random.randint(1, data.shape[0], 100)
select
```

```python
# select에 입력된 index를 기준으로 데이터 추출!

height_100 = data['키'][select]
weight_100 = data['몸무게'][select]

plt.scatter(weight_100, height_100)
plt.title("Korean teen health data")
plt.xlabel("weight")
plt.ylabel("height")
plt.show()
```

```python
# 기울기(a), y절편(b)인 직선의 함수 만들기.

def height_pred(x, a, b):
  y = a*x + b
  return(y)
```

```python
a = 0.8
b = 130

x = np.linspace(0, 100, 1001)
y = height_pred(x, a, b)

plt.scatter(weight_100, height_100)
plt.plot(x, y, 'r--')
plt.text(70, 130, "a : {}, b : {}".format(a, b))
plt.title("Korean teen health data")
plt.xlabel("weight")
plt.ylabel("height")

plt.xlim([0,100])
plt.ylim([100, 190])
plt.show()
```

```python
# Error를 구하기 위한 DataFrame을 df에 생성
# DataFrame끼리 합칠때는 concat을 사용한다.

df = pd.concat([weight_100, height_100], axis = 1)
df
```

```python
# index reset!

df.reset_index(drop = True, inplace = True)
df
```

```python
# linear function로 예측한 예상 키를 df에 첨가.

a = 0.8
b = 105

df["예상 키"] = height_pred(df["몸무게"], a, b)
df
```

```python
# 예상 키와 실체 키와의 차이를 게산하기.

diff = df['키'] - df['예상 키']
diff
```

```python
# RMSE 구하기 

np.sqrt(sum(diff**2)/len(diff))
```

```python
# a와 b에 따른 RMSE변화를 시각적으로 살펴보자!

a, b = 1.2, 130

plt.figure(figsize = (8,6))
x = np.linspace(0,100,1001)
y = height_pred(x, a, b)

df = pd.concat([weight_100, height_100], axis = 1)
df["예상 키"] = height_pred(df['몸무게'], a, b)
diff = df['키'] - df['예상 키']

RMSE = np.sqrt(sum(diff**2)/len(diff))

# 시각화

plt.scatter(weight_100, height_100)
plt.plot(x, y, 'r--')

plt.title("Korean teen health data", fontsize = 20)
plt.xlabel("weight")
plt.ylabel("height")
plt.text(70, 130, "a : {}, b : {}".format(a, b))
plt.text(71, 120, "RMSE : {:1.3f}".format(RMSE), fontsize = 15)

plt.xlim([0,100])
plt.ylim([100, 190])
plt.show()
```

```python
# 가능한 a, b 의 쌍을 모두 입력해보고 최소 RMSE를 뽑아보자!

data_rmse = pd.DataFrame([a, b, RMSE]).T
data_rmse.columns = ["a", "b", "RMSE"]
data_rmse
```

```python
a_range = np.linspace(0.9, 1.1, 3)
b_range = np.linspace(90, 110, 3)

a_list = []
b_list = []
rmse_list = []

df = pd.concat([weight_100, height_100], axis = 1)

for a in a_range:
  for b in b_range:
    df["예상 키"] = height_pred(df['몸무게'], a, b)
    diff = df['키'] - df['예상 키']
    RMSE = np.sqrt(sum(diff**2)/len(diff))

    a_list.append(a)
    b_list.append(b)
    rmse_list.append(RMSE)


data_rmse = pd.DataFrame([a_list, b_list, rmse_list]).T
data_rmse.columns = ['a', 'b', 'RMSE']
data_rmse
```

```python
x = np.linspace(0, 100, 1001)

for k in range(data_rmse.shape[0]):
  a, b, RMSE = data_rmse.iloc[k,:]

  y = height_pred(x, a, b)
  plt.scatter(weight_100, height_100)
  plt.plot(x, y, 'r--')

  plt.title("Korean teen health data", fontsize = 20)
  plt.xlabel("weight")
  plt.ylabel("height")
  plt.text(70, 130, "a : {}, b : {}".format(a, b))
  plt.text(71, 120, "RMSE : {:1.3f}".format(RMSE), fontsize = 15)

  plt.xlim([0,100])
  plt.ylim([100, 190])
  plt.show()
  print("\n")
```
